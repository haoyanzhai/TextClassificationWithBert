```python
import os
import pandas as pd
```

```python
with open('./reuters/reuters/reuters/cats.txt', 'r') as f:
    labels = f.read()
labels = labels.split('\n')
labels = [x.split(maxsplit=1) for x in labels]
train_label = {}
test_label = {}
for i, tmp in enumerate(labels):
    if len(tmp) < 2:
        print(i, tmp)
        continue
    title, label = tmp
    x = title.split('/')
    y = label.split()
    if len(y) > 1: # skip mult-label text
        continue
    if x[0] == 'test':
        test_label[x[1]] = y[0]
    else:
        train_label[x[1]] = y[0]
```

```python
# train
```

```python
df_dict = {'text': [], 'label': []}

directory = './reuters/reuters/reuters/training/'
for file in os.listdir(directory):
    dir_file = os.path.join(directory, file)
    if file in train_label:
        print(dir_file)
        with open(dir_file, 'r') as f:
            text = f.read()
            df_dict['text'].append(text)
            df_dict['label'].append(train_label[file])
df = pd.DataFrame(df_dict)
```

```python
label_mapping = {val: key for key, val in dict(enumerate(set(df['label']))).items()}
df['label1'] = df['label'].map(label_mapping)
```

```python
df.to_parquet('./reuters_train.parquet')
```

```python
# test
```

```python
df_dict = {'text': [], 'label': []}

directory = './reuters/reuters/reuters/test/'
for file in os.listdir(directory):
    dir_file = os.path.join(directory, file)
    if file in test_label:
        print(dir_file)
        with open(dir_file, 'r') as f:
            text = f.read()
            df_dict['text'].append(text)
            df_dict['label'].append(test_label[file])
df = pd.DataFrame(df_dict)
```

```python
df['label1'] = df['label'].map(label_mapping)
df = df.dropna(subset=['label1']).copy()
df['label1'] = df['label1'].astype(int)
```

```python
df.to_parquet('./reuters_test.parquet')
```
